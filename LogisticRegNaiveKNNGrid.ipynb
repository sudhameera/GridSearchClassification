```python
import pandas as pd
```

```python
data =pd.read_csv("Social_Network_Ads.csv")
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data["Purchased"].value_counts()
```

```python
data.drop("User ID",axis=1)
```

```python
data.columns
```

```python
independent=data[['User ID', 'Age', 'EstimatedSalary', 'Gender_Male']]
```

```python
dependent=data[[ 'Purchased']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

grid_params={"solver":['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag',' saga'],
             "multi_class":['auto', 'ovr', 'multinomial']}
grid=GridSearchCV(LogisticRegression(),grid_params,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')

grid.fit(x_train,y_train)
```

```python
results=grid.cv_results_
results
```

```python
table=pd.DataFrame.from_dict(results)
table
```

```python
print("The  f1_weighted value with best tuning parameter is{} :".format(grid.best_params_))
```

```python
ypred=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypred)
cm
```

```python
from sklearn.metrics import classification_report
class_report=classification_report(y_test,ypred)
print(class_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test) [:,1])
```
